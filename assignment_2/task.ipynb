```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def show_img(img, title=""):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
def padding(img, border_width):
    img = cv2.copyMakeBorder(img, 
                             border_width, 
                             border_width, 
                             border_width, 
                             border_width, 
                             cv2.BORDER_REFLECT)
    return img

def crop(img, x0, x1, y0, y1):
    return img[y0:y1, x0:x1]

def resize(img, width, height):
    return cv2.resize(img, (width, height))

def copy(img, arr):
    height, width, channels = img.shape
    arr = np.zeros((height, width, channels), dtype=np.uint8)
    arr[:] = img

    return arr

def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def hsv(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

def hue_shifted(img, arr, hue):
    height, width, channels = img.shape
    arr = np.zeros((height, width, channels), dtype=np.uint8)
    arr[:] = img + hue

    return arr

def smoothing(img):
    return cv2.GaussianBlur(img, (15, 15), cv2.BORDER_DEFAULT)

def rotation(img, angle):
    if angle == 90:
        return cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    elif angle == 180:
        return cv2.rotate(img, cv2.ROTATE_180)
    else:
        raise Exception("Angle must be 90 or 180")
```

```python
img = cv2.imread("lena-2.png")
height, width, _ = img.shape
copy_arr = None

show_img(img, "normal")
show_img(padding(img, 100), "padding")
show_img(crop(img, 80, width - 130, 80, height - 130), "crop")
show_img(resize(img, 200, 200), "resize")
show_img(copy(img, copy_arr), "copy")
show_img(grayscale(img), "grayscale")
show_img(hsv(img), "hsv")
show_img(hue_shifted(img, copy_arr, 50), "hue shifted")
show_img(smoothing(img), "smoothing")
show_img(rotation(img, 90), "rotate 90")
show_img(rotation(img, 180), "rotate 180")
```
